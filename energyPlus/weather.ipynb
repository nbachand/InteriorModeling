```python
%load_ext autoreload
%autoreload 2
%matplotlib widget

from epw import epw
from matplotlib import pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
a = epw()
a.read("./weather/USA_CA_Sacramento.724835_TMY2.epw")
a.dataframe.index = pd.to_datetime(a.dataframe[['Year', 'Month', 'Day', 'Hour', 'Minute']])
display(a.dataframe.keys())
display(set(a.dataframe["Year"]))

# radiation = a.dataframe[["Direct Normal Radiation"]]
# radiation = radiation.loc[(radiation!=0).any(axis=1)]
```

```python
# data = a.dataframe[(a.dataframe["Month"]==6) & (a.dataframe["Year"]==1985)]
# data = a.dataframe[a.dataframe["Year"]==1985]
data = a.dataframe[a.dataframe["Month"]==6]
display(f"Data for months: {set(data['Month'])}, years: {set(data['Year'])}")
radiation_keys = ["Extraterrestrial Horizontal Radiation",
    "Extraterrestrial Direct Normal Radiation" ,
    "Horizontal Infrared Radiation Intensity" ,
    "Global Horizontal Radiation" ,
    "Direct Normal Radiation" ,
    "Diffuse Horizontal Radiation" ]
fig = px.line(data, x = data.index, y = radiation_keys)
fig.show()
```

```python
data["Total Sky Radiation"] = data["Horizontal Infrared Radiation Intensity"] + data["Global Horizontal Radiation"]

fig = px.line(data, x = data.index, y = [
    "Horizontal Infrared Radiation Intensity",
    "Global Horizontal Radiation",
    "Total Sky Radiation",
    "Dry Bulb Temperature",
    ])
fig.show()
```
